<h1 style="color:red">Load Liberaries</h1>

```python
import joblib
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
from tensorflow.keras.models import load_model
```

<h1 style="color:red">Load Test data and its Labels</h1>

```python
x_test = joblib.load('pk1/xvalid_tfidf.pk1')
y_test=  joblib.load('pk1/valid_y.pk1')
```

<h1 style="color:red">Accuracy Calculation Template</h1>

```python
def accuracy_calc(model,model_name):
    predictions = model.predict(x_test)
    print(f"Accuracy For {model_name} is = {metrics.accuracy_score(predictions, y_test)}")
    #---------------Confussion Matrix------------------------
    cm = confusion_matrix(y_test, predictions)
    cm_percent = cm / cm.sum(axis=1, keepdims=True) * 100
    class_names = y_test.unique()
    plt.figure(figsize=(10, 10))
    sns.heatmap(cm_percent, annot=True, fmt=".1f", cmap="Blues")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.title("Confusion Matrix Heatmap")
    plt.xticks(np.arange(len(class_names)) + 0.5, class_names)
    plt.yticks(np.arange(len(class_names)) + 0.5, class_names)
    plt.show()
```

<h1 style="color:red">Naive Bayes Accuracy</h1>

```python
naive_bayes_model = joblib.load('pk1/Naive Bayes Multinominal.pk1')
accuracy_calc(naive_bayes_model ,"Naive Bayes Multinominal" )
#Accuracy For Naive Bayes Multinominal is = 0.8661710037174721
```

<h1 style="color:red">Logistic Regression Accuracy</h1>

```python
logistic_regression_model = joblib.load('pk1/Logistic Regression.pk1')
accuracy_calc(logistic_regression_model ,"Logistic Regression" )
#Accuracy For Logistic Regression is = 0.8629845990440786
```

<h1 style="color:red">Support Vector Machine Accuracy</h1>

```python
svm_model = joblib.load('pk1/Support Vector Classifier.pk1')
accuracy_calc(svm_model ,"Support Vector Classifier" )
#Accuracy For Support Vector Classifier is = 0.8757302177376527
```

<h1 style="color:red">Random Forest Accuracy</h1>

```python
RNF_model = joblib.load('pk1/Random Forest Classifier.pk1')
accuracy_calc(RNF_model ,"Random Forest Classifier" )
#Accuracy For Random Forest Classifier is = 0.8066914498141264
```

<h1 style="color:red">Neural network Accuracy</h1>

```python
NN_model = load_model("pk1/NN_model.h5")
x_test_reshaped = x_test.toarray()
loss, accuracy = NN_model.evaluate(x_test_reshaped, y_test)
print(f"Loss is  {loss}")
print(f"accuracy is {accuracy}")

prediction = NN_model.predict(x_test_reshaped)
prediction = np.argmax(prediction, axis=1)

cm = confusion_matrix(y_test, prediction)
cm_percent = cm / cm.sum(axis=1, keepdims=True) * 100
class_names = y_test.unique()
plt.figure(figsize=(10, 10))
sns.heatmap(cm_percent, annot=True, fmt=".1f", cmap="Blues")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix Heatmap")
plt.xticks(np.arange(len(class_names)) + 0.5, class_names)
plt.yticks(np.arange(len(class_names)) + 0.5, class_names)
plt.show()
```
